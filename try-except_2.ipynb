## Q4

The LookupError class and its derived exceptions are used to handle scenarios where data lookup operations fail due to missing or invalid keys, indices, or items in various data structures. These exceptions provide a way to gracefully handle such situations and provide meaningful error messages to users or developers.

1. IndexError

```python
x = [1,2,4]
try:
    result = x[10]
    print(result)
except IndexError as e:
    print(e)
```

2. KeyError

```python
my_dict = {'a': 1, 'b': 2}
try:
    value = my_dict['c']
except KeyError as e:
    print("Error:", e)
```
